```python
import sys
sys.path.append("../src/")
import pandas as pd
import numpy as np
from pyspark.sql import SparkSession
# create sparksession
spark = SparkSession \
    .builder \
    .appName("Pysparkexample") \
    .config("spark.some.config.option", "some-value") \
    .getOrCreate()
from config import *
```

```python
df = spark.read.csv("../data/census-income.data", header=False, inferSchema=True)
```

```python
df.schema
```

```python
df.limit(10).toPandas()
```

```python
df = df.drop("_c24")
```

```python
df.limit(10).toPandas()
```

```python
df_descr = pd.read_csv("../data/column_name.csv")
```

```python
columns = list(df_descr["Column Name"])
```

```python
len(columns)
```

```python
len(df.columns)
```

```python
df = df.toDF(*columns)
```

```python
df.schema
```

```python
df.count(), len(df.columns)
```

```python
df.summary().toPandas().set_index("summary").T
```

```python
df.groupBy("age").count().orderBy("count").toPandas()
```

```python
cols = []
uniques = []
for col in df.columns:
    unique = ", ".join([str(a[col]) for a in df.select(col).distinct().collect()])
    cols.append(col)
    uniques.append(unique)

pandas_df_unique_values = pd.DataFrame({"Var": cols, "Unique": uniques})
```

```python
pandas_df_unique_values
```

```python
pandas_df_unique_values.to_parquet("../reports/unique_values.parquet")
```
