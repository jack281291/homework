## Missing values

```python
def count_missings(spark_df,sort=True):
    df = spark_df.select([F.count(F.when(F.isnan(c) | F.isnull(c), c)).alias(c) for c in spark_df.columns]).toPandas()

    if len(df) == 0:
        print("There are no any missing values!")
        return None

    if sort:
        return df.rename(index={0: 'count'}).T.sort_values("count",ascending=False)

    return df
```

```python
count_missings(df)
```

## Categorical variables

```python
cat_vars = [col for col in df.columns if col not in cont_vars]
```

```python
cat_vars_0 = [col for col in cat_vars if len(dict_unique[col].split(",")) <= 20]
cat_vars_1 = [col for col in cat_vars if len(dict_unique[col].split(",")) > 20]
```

```python
len(cat_vars_0)
```

```python
len(cat_vars_1)
```

```python
for col in cat_vars_0:
    df.groupBy(col).count().toPandas().set_index(col).plot(kind='barh')
```

```python
for col in cat_vars_1:
    print(df.groupBy(col).count().orderBy("count", ascending=False).show())
```

**For a lot of categorical variables it makes sense to create groups, especially if we want to use a logistic regression**
